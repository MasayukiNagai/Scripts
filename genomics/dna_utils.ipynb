```python
from dna_utils import OneHotEncoder
```

```python
# by default, channel_axis=-1 (N, L, 4)
encoder = OneHotEncoder("ACGT")
sequences = [
    "ACGTACGTACGTACGTACGT",
    "AAAAAAAAAAAAAAAAAAAA",
    "CCCCCCCCCCCCCCCCCCCC",
    "GGGGGGGGGGGGGGGGGGGG",
    "TTTTTTTTTTTTTTTTTTTT",
    "NNNNNNNNNNNNNNNNNNNN",
    "ATATCGCGATATCGCGNNNA",
]
onehot = encoder.to_onehot(sequences)
print(onehot.shape)
decoded = encoder.from_onehot(onehot)
```

```python
for i, seq in enumerate(sequences):
    print(sequences[i])
    print(decoded[i])
    print(onehot[i])
```

```python
# with channel_axis=1 (N, 4, L)
encoder = OneHotEncoder("ACGT", channel_axis=1)
sequences = [
    "ACGTACGTACGTACGTACGT",
    "AAAAAAAAAAAAAAAAAAAA",
    "CCCCCCCCCCCCCCCCCCCC",
    "GGGGGGGGGGGGGGGGGGGG",
    "TTTTTTTTTTTTTTTTTTTT",
    "NNNNNNNNNNNNNNNNNNNN",
    "ATATCGCGATATCGCGNNNA",
]
onehot = encoder.to_onehot(sequences)
print(onehot.shape)
decoded = encoder.from_onehot(onehot)
```

```python
for i, seq in enumerate(sequences):
    print(sequences[i])
    print(decoded[i])
    print(onehot[i])
```

```python
# different lengths of sequences
encoder = OneHotEncoder("ACGT", channel_axis=1)
sequences = [
    "ACGTACGTACGT",
    "AAAAAAAAAAAAAAAA",
    "CCCCCCCCCCCCCCCCC",
    "GGGGGGGGGGGGGGGGGGGG",
    "TTTTTTTTTTTTTTTTTTTT",
    "NNNNNNNNNNNNNNNNNNNN",
    "ATATCGCGATATCGCGNNNA",
]
onehot = encoder.to_onehot(sequences)
print(onehot.shape)
decoded = encoder.from_onehot(onehot)

# short sequences gets padded with unknowns at the end
print(decoded)
```

```python
from dna_utils import reverse_complement, reverse_complement_string
```

```python
rc_sequences = [reverse_complement_string(s) for s in sequences]
rc_sequences
```

```python
rc_onehot = reverse_complement(onehot)
print(rc_onehot.shape)
rc_decoded = encoder.from_onehot(rc_onehot)
print(rc_decoded)
```
